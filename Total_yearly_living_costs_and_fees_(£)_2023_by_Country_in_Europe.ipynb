```python
df.sort_values(by='Total yearly living costs and fees (£)', ascending=False, inplace=True)
```

```python
lista_etiquetas = df['Country'].tolist()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
val = df['Total yearly living costs and fees (£)']

fig, ax = plt.subplots(figsize=(12, 6), dpi=85)
sns.barplot(x=lista_etiquetas, y=val, hue=lista_etiquetas, palette="RdBu", legend=False, ax=ax)
plt.xticks(rotation=90, fontsize='x-small')
ax.set_ylabel('Total yearly living costs and fees (£) 2023')
ax.set_xlabel('Country')
ax.set_title('Total yearly living costs and fees (£) 2023 by Country in Europe', fontweight='bold', fontsize='15')

plt.show()
```
